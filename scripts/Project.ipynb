```python
from ast import literal_eval
import numpy as np
import pandas as pd
```

```python
mcdf = pd.read_csv("../data/raw.csv", sep='|')
mcdf["userscore"] = pd.to_numeric(mcdf["userscore"], errors="coerce")
mcdf = mcdf.dropna()  # Removes rows missing any data
mcdf["critics"] = mcdf["critics"].apply(literal_eval)
mcdf["genres"] = mcdf["genres"].apply(literal_eval)
```

```python
expanded_critics = pd.json_normalize(mcdf["critics"]).set_index(mcdf.index)
mcdf = pd.concat([mcdf, expanded_critics], axis=1).drop("critics", axis=1)
```

```python
exploded = mcdf["genres"].explode().where(lambda x: x != '').dropna()
count = exploded.value_counts()
# Keep most common genre labels
prevalent = exploded[exploded.isin(count[count > 85].index)]
ohe_genres = pd.crosstab(prevalent.index, prevalent)
pd.concat([mcdf, ohe_genres], axis=1).drop("genres", axis=1)
```
